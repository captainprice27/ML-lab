## **Regularization Techniques: L1, L2, and None**

***Effect:*** L2 regularization does not produce sparse models, but it reduces the magnitude of coefficients, leading to more generalized models.    
## **3. None (No Regularization):**

***Definition:*** In the context of logistic regression, "None" means no regularization is applied. The model purely tries to minimize the loss without any penalty on the coefficients.
***Effect:*** Without regularization, the model might overfit the training data, especially if the data is noisy or if there are many features.

## **Solvers in Logistic Regression**
**1. Newton-CG (Newton-Conjugate Gradient):**

***Description:*** Newton-CG is an optimization algorithm that uses the Hessian matrix (second-order derivative) to find the minimum of the loss function. It's efficient for large datasets but only supports L2 regularization or no regularization.   
***Best Used For:*** Problems with large datasets where L2 regularization is sufficient. Itâ€™s particularly efficient for multi-class classification problems.    
**2. LBFGS (Limited-memory Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno):**

***Description:*** LBFGS is an optimization algorithm that approximates the Hessian matrix without storing it, making it memory efficient. Like Newton-CG, it only supports L2 regularization or no regularization.
***Best Used For:*** Large datasets and multi-class problems, where memory efficiency is crucial.    
**3. Liblinear:**   
***Description:*** Liblinear is an implementation of linear models for large datasets. Itâ€™s based on coordinate descent algorithms and supports both L1 and L2 regularization.    
***Best Used For:*** Smaller datasets or problems where feature selection is important, as it supports L1 regularization.   



| Solver    | L1 Support | L2 Support | None Support | Use Cases                                                                 |

| LBFGS     | âœ˜          | âœ”          | âœ”            | Memory-efficient, ideal for large datasets, supports only L2 or None.     |
| Liblinear | âœ”          | âœ”          | âœ˜            | Supports both L1 and L2, ideal for smaller datasets or when feature selection (L1) is needed. |


## a.) Use â€˜**newton-cg**â€™, â€˜**lbfgs**â€™, â€˜**liblinear**â€™ solver for the regression model.
## b.) For each solver use **â€˜L1â€™**, **â€˜L2â€™**, **â€˜Noneâ€™** penalty to train the model.

```python
solvers_with_none = ['newton-cg', 'lbfgs']
solvers_l2_only = ['newton-cg', 'lbfgs']
solvers_l1_l2 = ['liblinear']
penalties_l2_none = ['l2', None]
penalties_l1_l2 = ['l1', 'l2']

# Initialize a DataFrame to store the coefficients
coefficients_df = pd.DataFrame(columns=[f'Feature_{i}' for i in range(1, 31)])

# Train models with solvers that support 'l2' and 'none' penalties
for solver in solvers_l2_only:
    for penalty in penalties_l2_none:
        if penalty is None:
            model = LogisticRegression(solver=solver, penalty='none', max_iter=10000)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)

        model.fit(X_train, y_train)
        coefficients_df.loc[f'{solver}_{penalty}'] = model.coef_[0]

# Train models with the 'liblinear' solver (supports 'l1' and 'l2' only)
for solver in solvers_l1_l2:
    for penalty in penalties_l1_l2:
        model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)
        model.fit(X_train, y_train)
        coefficients_df.loc[f'{solver}_{penalty}'] = model.coef_[0]

# Display the coefficients table
# print(coefficients_df)

print("Coefficients Table:")
print(coefficients_df.round(4))  # Round to 4 decimal places for readability
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure
plt.figure(figsize=(32, 16))

# Create a heatmap of the coefficients
sns.heatmap(coefficients_df, annot=True, cmap='coolwarm', fmt='.4f', linewidths=0.5)

# Add titles and labels
plt.title('Heatmap of Coefficients of all 30 features')
plt.xlabel('Features')
plt.ylabel('Solver_Penalty')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

```python
# Evaluate models with solvers that support 'l2' and 'none' penalties
for solver in solvers_l2_only:
    for penalty in penalties_l2_none:
        if penalty is None:
            model = LogisticRegression(solver=solver, penalty='none', max_iter=100)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=100)

        model.fit(X_train, y_train)

        # Predict on the test set
        y_pred = model.predict(X_test)

        # Print accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f'Accuracy for {solver} with {penalty} penalty: {accuracy:.4f}')

# Evaluate models with the 'liblinear' solver (supports 'l1' and 'l2')
for solver in solvers_l1_l2:
    for penalty in penalties_l1_l2:
        model = LogisticRegression(solver=solver, penalty=penalty, max_iter=100)
        model.fit(X_train, y_train)

        # Predict on the test set
        y_pred = model.predict(X_test)

        # Print accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f'Accuracy for {solver} with {penalty} penalty: {accuracy:.4f}')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define the solvers and penalties
solvers_with_none = ['newton-cg', 'lbfgs']
solvers_l2_only = ['newton-cg', 'lbfgs']
solvers_l1_l2 = ['liblinear']
penalties_l2_none = ['l2', None]
penalties_l1_l2 = ['l1', 'l2']

print("Before Fine-Tuning:")
# Evaluate models with solvers that support 'l2' and 'none' penalties
for solver in solvers_l2_only:
    for penalty in penalties_l2_none:
        if penalty is None:
            model = LogisticRegression(solver=solver, penalty='none', max_iter=10000)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)

        # Train on the initial 80% training data
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, pos_label='M')  # Assuming 'M' is the positive label
        recall = recall_score(y_test, y_pred, pos_label='M')
        f1 = f1_score(y_test, y_pred, pos_label='M')

        print(f'Metrics for {solver} with {penalty} penalty:')
        print(f'  Accuracy: {accuracy:.4f}')
        print(f'  Precision: {precision:.4f}')
        print(f'  Recall: {recall:.4f}')
        print(f'  F1-score: {f1:.4f}')

# Evaluate models with the 'liblinear' solver (supports 'l1' and 'l2')
for solver in solvers_l1_l2:
    for penalty in penalties_l1_l2:
        model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, pos_label='M')
        recall = recall_score(y_test, y_pred, pos_label='M')
        f1 = f1_score(y_test, y_pred, pos_label='M')

        print(f'Metrics for {solver} with {penalty} penalty:')
        print(f'  Accuracy: {accuracy:.4f}')
        print(f'  Precision: {precision:.4f}')
        print(f'  Recall: {recall:.4f}')
        print(f'  F1-score: {f1:.4f}')

# After Fine-Tuning
print("\nAfter Fine-Tuning:")
for solver in solvers_l2_only:
    for penalty in penalties_l2_none:
        if penalty is None:
            model = LogisticRegression(solver=solver, penalty='none', max_iter=10000)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)

        # Fine-tune using validation data
        model.fit(X_val, y_val)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, pos_label='M')
        recall = recall_score(y_test, y_pred, pos_label='M')
        f1 = f1_score(y_test, y_pred, pos_label='M')

        print(f'Metrics for {solver} with {penalty} penalty:')
        print(f'  Accuracy: {accuracy:.4f}')
        print(f'  Precision: {precision:.4f}')
        print(f'  Recall: {recall:.4f}')
        print(f'  F1-score: {f1:.4f}')

# Evaluate models with the 'liblinear' solver after fine-tuning
for solver in solvers_l1_l2:
    for penalty in penalties_l1_l2:
        model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)

        # Fine-tune using validation data
        model.fit(X_val, y_val)
        y_pred = model.predict(X_test)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, pos_label='M')
        recall = recall_score(y_test, y_pred, pos_label='M')
        f1 = f1_score(y_test, y_pred, pos_label='M')

        print(f'Metrics for {solver} with {penalty} penalty:')
        print(f'  Accuracy: {accuracy:.4f}')
        print(f'  Precision: {precision:.4f}')
        print(f'  Recall: {recall:.4f}')
        print(f'  F1-score: {f1:.4f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Select 5 random features
np.random.seed(71)  # for reproducibility
selected_features = np.random.choice(range(30),6, replace=False)

# Initialize dictionaries to store coefficients
coefficients_before = {f'Feature_{i+1}': [] for i in selected_features}
coefficients_after = {f'Feature_{i+1}': [] for i in selected_features}

# Before Fine-Tuning
for solver in solvers_l2_only:
    for penalty in penalties_l2_none:
        if penalty is None:
            model = LogisticRegression(solver=solver, penalty='none', max_iter=10000)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)

        model.fit(X_train, y_train)
        for i in selected_features:
            coefficients_before[f'Feature_{i+1}'].append(model.coef_[0][i])

for solver in solvers_l1_l2:
    for penalty in penalties_l1_l2:
        model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)
        model.fit(X_train, y_train)
        for i in selected_features:
            coefficients_before[f'Feature_{i+1}'].append(model.coef_[0][i])

# After Fine-Tuning
for solver in solvers_l2_only:
    for penalty in penalties_l2_none:
        if penalty is None:
            model = LogisticRegression(solver=solver, penalty='none', max_iter=10000)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)

        model.fit(X_val, y_val)
        for i in selected_features:
            coefficients_after[f'Feature_{i+1}'].append(model.coef_[0][i])

for solver in solvers_l1_l2:
    for penalty in penalties_l1_l2:
        model = LogisticRegression(solver=solver, penalty=penalty, max_iter=10000)
        model.fit(X_val, y_val)
        for i in selected_features:
            coefficients_after[f'Feature_{i+1}'].append(model.coef_[0][i])

# Plot the coefficients
plt.figure(figsize=(14, 8))

for i, feature in enumerate(selected_features):
    plt.subplot(2, 3, i+1)
    plt.plot(coefficients_before[f'Feature_{feature+1}'], label='Before Fine-Tuning', marker='o')
    plt.plot(coefficients_after[f'Feature_{feature+1}'], label='After Fine-Tuning', marker='o')
    plt.title(f'Feature {feature+1}')
    plt.xlabel('Solver-Penalty Combinations')
    plt.ylabel('Coefficient')
    plt.xticks(ticks=range(len(coefficients_before[f'Feature_{feature+1}'])),
               labels=['newton-cg-l2', 'newton-cg-none', 'lbfgs-l2', 'lbfgs-none', 'liblinear-l1', 'liblinear-l2'],
               rotation=45, ha='right')
    plt.legend()

plt.tight_layout()
plt.show()
```
